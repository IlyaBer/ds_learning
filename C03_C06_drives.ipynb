```python
import pandas as pd
import plotly
import plotly.express as px
speeds = pd.read_excel('C:/Users/ilia.berdan/Downloads/speed_C03_C06_3_3_2_2_2.xlsx')
```

```python
a=13
b=0
speeds06=speeds

try:
    while a!=31:
        speeds=speeds.drop(index=a)
        a=a+1
except KeyError:
    print()
 
try:
    while b!=15:
        speeds06=speeds06.drop(index=b)
        b+=1
except KeyError:
    print()
    
#speeds06   
#speeds
```

```python
new=speeds.T

col_ind=list()
drive_id=list()
for i in range(0,13):
    col_ind.append(new.loc['date'][i])
    drive_id.append(new.loc['date'][i])

col_ind.reverse()
col_ind.append('date')
col_ind.reverse()

new=new.reset_index()
new.columns=col_ind
new=new.drop([0])
#new.head()

##########################

new06=speeds06.T

col_ind06=list()
drive_id06=list()

for f in range(15,29):
    col_ind06.append(new06.loc['date'][f])
    drive_id06.append(new06.loc['date'][f])

col_ind06.reverse()
col_ind06.append('date')
col_ind06.reverse()

new06=new06.reset_index()
new06.columns=col_ind06
new06=new06.drop([0])
#new06.head()
```

```python
cum_metric=pd.DataFrame()

new_num=new.apply(lambda col:pd.to_numeric(col, errors='coerce'))
new_num['date']=new['date']
cum_metric['date']=new['date']

cum_ind=list()

for elem in drive_id:
    cum_metric['cum_'+elem]=new_num[elem].cumsum()


mes=1
for item in cum_metric.iterrows():
    cum_metric.at[mes,'measures']=mes
    mes+=1
#cum_metric.head()

###########################

cum_metric06=pd.DataFrame()

new_num06=new06.apply(lambda col:pd.to_numeric(col, errors='coerce'))
new_num06['date']=new06['date']
cum_metric06['date']=new06['date']

cum_ind06=list()

for elem in drive_id06:
    cum_metric06['cum_'+elem]=new_num06[elem].cumsum()


mes06=1
for item in cum_metric06.iterrows():
    cum_metric06.at[mes06,'measures']=mes06
    mes06+=1
#cum_metric06.head()
```

```python
cmtr_names=list()
for elem in drive_id:
    cum_metric['cmtr_'+elem]=(cum_metric['cum_'+elem]/cum_metric['measures'])
    cum_metric=cum_metric.drop(['cum_'+elem], axis=1)
    cmtr_names.append('cmtr_'+elem)

#cum_metric.head()
#cmtr_names

#############

cmtr_names06=list()
for elem in drive_id06:
    cum_metric06['cmtr_'+elem]=(cum_metric06['cum_'+elem]/cum_metric06['measures'])
    cum_metric06=cum_metric06.drop(['cum_'+elem], axis=1)
    cmtr_names06.append('cmtr_'+elem)
    
#cum_metric06.head()
#cmtr_names06
```

```python

fig01 = px.line(
    data_frame=cum_metric, #DataFrame
    x='date', # 
    y=cmtr_names, # 
    height=700, #
    width=1500, #
    title='   C03()', #
    markers=True
)
fig01.show()

line_data = new.groupby('date', as_index=False).sum()
fig02 = px.line(
    data_frame=line_data, #DataFrame
    x='date', # 
    y=drive_id, # 
    height=700, #
    width=1500, #
    title='  C03(. )' #
)
fig02.show()
```

```python
fig03 = px.line(
    data_frame=cum_metric06, #DataFrame
    x='date', # 
    y=cmtr_names06, # 
    height=700, #
    width=1500, #
    title='   C06()', #
    markers=True
)
fig03.show()

line_data06 = new06.groupby('date', as_index=False).sum()
fig04 = px.line(
    data_frame=line_data06, #DataFrame
    x='date', # 
    y=drive_id06, # 
    height=700, #
    width=1500, #
    title='  C06(. )' #
)
fig04.show()
```

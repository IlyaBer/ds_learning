Для начала обсудим условия решения задачи:

    * Опыт работы 3 года 2 месяца…

В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.


```python
#ваш код здесь
mask=modified_df['Опыт работы'].values =='Не указано'
modified_df.loc[mask, 'Опыт работы']=np.nan


def get_experience(arg):                # перерабатываем годы и месяцы опыта в месяцы
    if arg is np.nan:
        return np.nan
    year_month=arg.split(' ')
    year=0
    month=0
    if year_month[3]=='месяца' or year_month[3]=='месяцев' or year_month[3]=='месяц':
        month=year_month[2]
    else:
        year= year_month[2]
        if year_month[5]=='месяца' or year_month[5]=='месяцев' or year_month[5]=='месяц':
            month=year_month[4]
    
    year=int(year)
    month=int(month)    
    res=year*12+month
    return res

modified_df['Опыт работы (месяц)']=modified_df['Опыт работы'].apply(get_experience)

print('Медианное значение опыта работы среди соискателей: ')
display(modified_df['Опыт работы (месяц)'].median())

modified_df.info()
modified_df=modified_df.drop('Опыт работы', axis=1)

```

Создадим отдельные признаки **"Город"**, **"Готовность к переезду"**, **"Готовность к командировкам"**. При этом важно учесть:

    * … , готов к переезду , …

    * … , хочу переехать (США) , …
    
    Нас интересует только сам факт возможности или желания переезда.
* Признак **"Готовность к командировкам"** должен иметь два возможных варианта: True или False. Обратите внимание, что возможны несколько вариантов описания готовности к командировкам в признаке "Город, переезд, командировки". Например:
    * … , готов к командировкам , … 
    * … , готова к редким командировкам , …

    Еще один важный факт: при выгрузки данных у некоторых соискателей "потерялась" информация о готовности к командировкам. Давайте по умолчанию будем считать, что такие соискатели не готовы к командировкам.
    
Выполните преобразования и удалите столбец **"Город, переезд, командировки"** из таблицы.

*Совет: обратите внимание на то, что структура текста может меняться в зависимости от указания ближайшего метро. Учите это, если будете использовать порядок слов в своей программе.*


```python
#ваш код здесь
def get_city(arg):
    city=arg.split(' ')
    city=city[0]
    million_cities = ['Новосибирск', 'Екатеринбург', 'Нижний Новгород', 'Казань', 'Челябинск', 'Омск', 'Самара', 'Ростов-на-Дону', 'Уфа', 'Красноярск', 'Пермь', 'Воронеж', 'Волгоград' ]
    if city in million_cities:
        city='Город-миллионник'
    elif city=='Москва' or city=='Санкт-Петербург':
        return city
    else:
        city='Другие'
    return city

def get_if_ready_for_move(arg):
    ready_to_move=arg.split(',')
    ready=ready_to_move[1]
    ready=ready.split(' ')
    if 'не' in ready:
        return False
    elif 'готов' in ready:
        return True
    elif 'готова' in ready:
        return True
    elif 'хочу' in ready:
        return True
    else:
        ready=ready_to_move[2]
        ready=ready.split(' ')
        if 'не' in ready:
            return False
        elif 'готов' or 'готова' or 'хочу' in ready:
            return True
#    return ready_to_move[:3]
            
    

def get_if_ready_for_trip(arg):
    ready_to_trip=arg.split(',')
    trip=ready_to_trip[-1]
    trip=trip.split(' ')
    if len(trip)==4:
        return True
    elif 'редким' in trip:
        return True
    else:
        return False
    
modified_df['Город']=modified_df['Город, переезд, командировки'].apply(get_city)
modified_df['Готовность к переезду']=modified_df['Город, переезд, командировки'].apply(get_if_ready_for_move)
modified_df['Готовность к командировкам']=modified_df['Город, переезд, командировки'].apply(get_if_ready_for_trip)
print('Процентное распределение соискателей по городам:')
display(modified_df['Город'].value_counts(normalize=True)*100)

mask_1=modified_df['Готовность к переезду']==True
ready_for_all=modified_df[mask_1]['Готовность к командировкам'].value_counts().loc[True]
result=ready_for_all/44744*100
print(f'Процентная доля соискателей, готовых к командировкам и переезду: {result.round()}%')

modified_df=modified_df.drop('Город, переезд, командировки', axis=1)
```

5. Рассмотрим поближе признаки **"Занятость"** и **"График"**. Сейчас признаки представляют собой набор категорий желаемой занятости (полная занятость, частичная занятость, проектная работа, волонтерство, стажировка) и желаемого графика работы (полный день, сменный график, гибкий график, удаленная работа, вахтовый метод).
На сайте hh.ru соискатель может указывать различные комбинации данных категорий, например:
* полная занятость, частичная занятость
* частичная занятость, проектная работа, волонтерство
* полный день, удаленная работа
* вахтовый метод, гибкий график, удаленная работа, полная занятость

Такой вариант признаков имеет множество различных комбинаций, а значит множество уникальных значений, что мешает анализу. Нужно это исправить!

Давайте создадим признаки-мигалки для каждой категории: если категория присутствует в списке желаемых соискателем, то в столбце на месте строки рассматриваемого соискателя ставится True, иначе - False.

Такой метод преобразования категориальных признаков называется One Hot Encoding и его схема представлена на рисунке ниже:
<img src=https://raw.githubusercontent.com/AndreyRysistov/DatasetsForPandas/main/ohe.jpg>
Выполните данное преобразование для признаков "Занятость" и "График", ответьте на контрольные вопросы, после чего удалите их из таблицы

```python
#ваш код здесь

zanyat=modified_df['Занятость'].copy()
schedule=modified_df['График'].copy()


def get_zanyat(arg):
    zan=arg.split(',')
    full=False
    partly=False
    project=False
    volunteering=False
    internship=False
    if 'полная занятость' in zan:
        full=True
    elif ' полная занятость' in zan:
        full=True
    else:
        full=False
        
    if 'частичная занятость' in zan:
        partly=True
    elif ' частичная занятость' in zan:
        partly=True
    else:
        partly=False
        
    if 'проектная работа'in zan:
        project=True
    elif ' проектная работа' in zan:
        project=True
    else:
        project=False
        
    if 'волонтерство'in zan:
        volunteering=True
    elif ' волонтерство' in zan:
        volunteering=True
    else:
        volunteering=False
        
    if 'стажировка' in zan:
        internship=True
    elif ' стажировка' in zan:
        internship=True
    else:
        internship=False
    
    return pd.Series([full, partly, project, volunteering, internship])
#    return partly
columns_busy={0:'полная занятость', 1:'частичная занятость', 2:'проектная работа', 3:'волонтерство', 4:'стажировка'}
busy=zanyat.apply(get_zanyat).rename(columns=columns_busy)
df_with_busy=pd.concat([modified_df, busy], axis=1)

def get_schedule(arg):
    sched=arg.split(',')
    long_trip=False
    flexible=False
    full_day=False
    scheduled=False
    remote_work=False
    if 'вахтовый метод' in sched:
        long_trip=True
    elif ' вахтовый метод' in sched:
        long_trip=True
    else:
        long_trip=False
        
    if 'гибкий график' in sched:
        flexible=True
    elif ' гибкий график' in sched:
        flexible=True
    else:
        flexible=False
        
    if 'полный день'in sched:
        full_day=True
    elif ' полный день' in sched:
        full_day=True
    else:
        full_day=False
        
    if 'сменный график'in sched:
        scheduled=True
    elif ' сменный график' in sched:
        scheduled=True
    else:
        scheduled=False
        
    if 'удаленная работа' in sched:
        remote_work=True
    elif ' удаленная работа' in sched:
        remote_work=True
    else:
        remote_work=False
    
    return pd.Series([long_trip, flexible, full_day, scheduled, remote_work])
columns_sched={0:'вахтовый метод', 1:'гибкий график', 2:'полный день', 3:'сменный график', 4:'удаленная работа'}
sched_df=schedule.apply(get_schedule).rename(columns=columns_sched)

df_sch_busy= pd.concat([df_with_busy, sched_df], axis=1)

print('Готовы к волонтерству и к проектной работе: ')
display(df_sch_busy[df_sch_busy['волонтерство']==True]['проектная работа'].value_counts())
print('Готовы к вахтовой работе и гибкому графику:')
display(df_sch_busy[df_sch_busy['вахтовый метод']==True]['гибкий график'].value_counts())

df_sch_busy=df_sch_busy.drop(['Занятость', 'График'], axis=1)
```

2. Выделить из столбца "ЗП" сумму желаемой заработной платы и наименование валюты, в которой она исчисляется. Наименование валюты перевести в стандарт ISO согласно с таблицей выше.

Значение close для рубля заполнить единицей 1 (курс рубля самого к себе)
4. Умножить сумму желаемой заработной платы на присоединенный курс валюты (close) и разделить на пропорцию (обратите внимание на пропуски после объединения в этих столбцах), результат занести в новый столбец "ЗП (руб)".


```python
#ваш код здесь

money=pd.read_csv('/Users/localadm/Saved Games/IDE/ds_learning/project_1/data/ExchangeRates.csv')

df_sch_busy['Обновление резюме']=pd.to_datetime(df_sch_busy['Обновление резюме'], dayfirst=True).dt.date
df_sch_busy['date']=df_sch_busy['Обновление резюме']
df_sch_busy['date']=pd.to_datetime(df_sch_busy['date'], dayfirst=True)
money['date']=pd.to_datetime(money['date'], dayfirst=True)
money=money.drop(['time', 'per', 'vol'], axis=1)

def get_money(arg):
    money=arg.split(' ')
    mon_amount=int(money[0])
    currency=money[1]
    
    return pd.Series([mon_amount, currency])

df_sch_busy[['Цифровое значение ЗП','currency']]=df_sch_busy['ЗП'].apply(get_money)
BYN_mask=df_sch_busy['currency']=='бел.руб.'
UAH_mask=df_sch_busy['currency']=='грн.'
UZS_mask=df_sch_busy['currency']=='сум'
df_sch_busy.loc[BYN_mask, 'currency']= 'BYN'
df_sch_busy.loc[UAH_mask, 'currency']= 'UAH'
df_sch_busy.loc[UZS_mask, 'currency']= 'UZS'
df_sch_busy['currency'].value_counts()


result_df=df_sch_busy.merge(
    money,
    on=['date', 'currency'],
    how='left'
)

result_df['close']=result_df['close'].fillna(1)
result_df['proportion']=result_df['proportion'].fillna(1)
course=result_df['close']
prop=result_df['proportion']
result_df['ЗП (руб.)']=result_df['Цифровое значение ЗП']/prop*course
result_df=result_df.drop(['Цифровое значение ЗП', 'currency', 'close', 'proportion','ЗП','date'], axis=1)
result_df['ЗП (руб.)'].median()

```

Есть ли аномалии для признака ЗП? Обратите внимание на гигантские размеры желаемой заработной платы.
*Совет: постройте гистограмму и коробчатую диаграмму рядом.*


```python
# ваш код здесь

def outliers_iqr(data, feature):           # функция, вычисляющая квантили и значения, лежащие за пределами максимальных границ коробчатой диаграммы(выбросы)
    x = data[feature]
    quartile_1, quartile_3 = x.quantile(0.25), x.quantile(0.75),
    iqr = quartile_3 - quartile_1
    lower_bound = quartile_1 - (iqr*2.5)
    upper_bound = quartile_3 + (iqr*2.5)
    outliers = data[(x<lower_bound) | (x > upper_bound)]
    cleaned = data[(x>lower_bound) & (x < upper_bound)]
    return outliers, cleaned

outliers, cleaned = outliers_iqr(result_df, 'ЗП (руб.)')

    
salary_data=cleaned['ЗП (руб.)']
#display(salary_data)
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(30,8))

axes[0].hist(
    x=salary_data,
    label=['ЗП (руб.)'],
    bins=40
)
axes[0].set_title('Распределение количества соискателей по желаемой ЗП')
axes[0].set_xlabel('Желаемая ЗП(руб.)')
axes[0].set_ylabel('Кол-во соискателей')
axes[0].grid()
axes[0].set_xticks(range(0, 250000, 10000))
axes[0].xaxis.set_tick_params(rotation=30);

axes[1].boxplot(
    salary_data
,    
    widths=2,
    patch_artist=True,
    showmeans=True,
    vert=0,
    medianprops={"color": "black", "linewidth": 0.5},
    boxprops={"facecolor": "C3", "edgecolor": "black",
                          "linewidth": 0.5},
)
axes[1].set_xticks(range(0, 250000, 10000))
axes[1].grid()
axes[1].set_title('Диапазон искомой ЗП')
axes[1].set_xlabel('ЗП(руб.)')
axes[1].xaxis.set_tick_params(rotation=30);
```

```python
# ваш код здесь


fig, aks=plt.subplots(1,1, figsize=(8, 4))
log_age=np.log(result_df['Возраст'] + 1)
histogr=sns.histplot(log_age, bins=20, ax=aks)
histogr.axvline(log_age.mean(), color='r', lw=2)
histogr.axvline(log_age.mean() - 3*log_age.std(), color='r', ls='--', lw=2)
histogr.axvline(log_age.mean() + 3*log_age.std(), color='r', ls='--', lw=2)
histogr.set_title('Возраст. Логарифмический масштаб');

def outliers_z_score(data, feature, log_scale=True):
    if log_scale:
        x = np.log(data[feature]+1)
    else:
        x = data[feature]
    mu = x.mean()           #   среднее
    sigma = x.std()         #   стандартное отклонение
    lower_bound = mu - 3 * sigma
    upper_bound = mu + 4 * sigma
    outliers = data[(x < lower_bound) | (x > upper_bound)]
    cleaned = data[(x > lower_bound) & (x < upper_bound)]
    return outliers, cleaned

outliers, cleaned = outliers_z_score(result_df, 'Возраст', log_scale=True)
display(outliers)

```
